```python
import numpy as np
import matplotlib.pyplot as plt

from lib.game import Game
```

## Game Simulation

```python
# Start a game
game = Game(
    num_players=3, 
    num_decks=6
    )

# Simulate data to array of shape (sims, rounds, players)
simdata, count_data = game.simulate_game(rounds=500, sims=1000)
```

```python
game.plot_simdata(player_index=1, simdata=simdata, log_scale=False, confidence=95)
game.plot_simdata(player_index=2, simdata=simdata, log_scale=False, confidence=95)
game.plot_simdata(player_index=0, simdata=simdata, log_scale=False, confidence=95)

# Plot a run of counts
rounds = count_data.shape[1]
x = np.arange(rounds)

fig, ax = plt.subplots(figsize=(16,3))
ax.plot(x, count_data[0, :, 0], label='Running Count', alpha=0.25)
ax.plot(x, count_data[0, :, 1], label='Total Count')
ax.axhline(0, color='grey', alpha=0.5)

ax.legend()

plt.show()
```

```python
# Compare round from counter vs non counter
round = 100
plt.hist(simdata[:, round, 0], density=True, bins='auto', alpha=0.5, label='Card Counter')
plt.hist(simdata[:, round, 2], density=True, bins='auto', alpha=0.5, label='Proportional Bets')
#plt.hist(simdata[:, round, 1], density=True, bins='auto', alpha=0.5, label='Constant Bets')

plt.xlim(0, np.percentile(simdata[:, round, 0], 99))

plt.legend()
plt.show()
```
